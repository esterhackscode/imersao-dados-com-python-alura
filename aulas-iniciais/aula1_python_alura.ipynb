```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
df.columns
```

```python
df = df.rename(columns={
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
})

display(df.head())
```

```python
df['senioridade'].value_counts()

# SE	Senior	Profissional experiente (nvel snior)
# MI	Mid	Nvel intermedirio
# EN	Entry	Iniciante ou jnior (entry-level)
# EX	Executive	Executivo, liderana snior (C-Level)
```

```python
df['contrato'].value_counts()

# FT	Full-time	Tempo integral  trabalho padro de 40h/semana ou perodo completo
# PT	Part-time	Meio perodo  carga horria reduzida
# CT	Contract	Contrato temporrio ou por projeto  geralmente por prazo determinado
# FL Freelance Freelancer  trabalho autnomo ou por demanda, sem vnculo formal
```

```python
df['remoto'].value_counts()

# 0 - vagas presenciais
# 100 - vagas remoto
# 50 - vagas hbridas
```

```python
df['tamanho_empresa'].value_counts()
```

```python
senioridade = {
    'SE': 'senior',
    'MI': 'pleno',
    'EN': 'junior',
    'EX': 'executivo'
}
df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
contrato = {
    'FT': 'integral',
    'PT': 'parcial',
    'CT': 'contrato',
    'FL': 'freelancer'
}
df['contrato'] = df['contrato'].replace(contrato)
df['contrato'].value_counts()
```

```python
tamanho_empresa = {
    'L': 'grande',
    'S': 'pequena',
    'M':	'media'

}
df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
mapa_trabalho = {
    0: 'presencial',
    100: 'remoto',
    50: 'hibrido'
}

df['remoto'] = df['remoto'].replace(mapa_trabalho)
df['remoto'].value_counts()
```
